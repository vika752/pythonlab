```python
import pygame
from pygame.draw import *
pygame.init()
clock = pygame.time.Clock()

FPS = 30
screen = pygame.display.set_mode((400, 400))

polygon(screen, (255, 255, 255), [(0,0), (0,400),
                       (400,400), (400,0)])
circle(screen, (255, 185, 0), (200, 175), 150)
polygon(screen, (0, 0, 0), [(220, 105), (240, 150),
                            (350,120), (370,75)])

circle(screen, (255, 0, 0), (280, 175), 30)
circle(screen, (0, 0, 0), (280, 175), 15)

circle(screen, (255, 0, 0), (120, 175), 30)
circle(screen, (0, 0, 0), (120, 175), 15)

polygon(screen, (0, 0, 0), [(50, 50), (40, 60),
                           (140, 160), (150, 150)])

polygon(screen, (0, 0, 0), [(150, 250), (150, 300),
                           (250, 300), (250, 250)])

pygame.display.update()
clock = pygame.time.Clock()
finished = False

while not finished:
    clock.tick(FPS)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            finished = True

pygame.quit()
```

```python
import pygame
from pygame.draw import *
pygame.init()
clock = pygame.time.Clock()
FPS = 30
screen = pygame.display.set_mode((600, 900))
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
GRAY = (205, 205, 205)
DARK_GRAY = (105, 105, 105)
BLUE = (0, 0, 255)
LIGHT_BLUE = (64, 128, 255)
GREEN = (0, 200, 64)
RED = (255, 0, 0)
YELLOW = (225, 225, 0)
PINK = (230, 50, 230)

def opacity(a, x, y):
    s=pygame.Surface([x, y])
    s.fill(BLACK)
    s.set_colorkey(BLACK)
    s.set_alpha(int(255*a))
    return s

def fish(a, x, y, color, scr):
    if(a<0):
        d1=int(60*abs(a))
        c=-1
    else: 
        c=1
        d1=0
    print(x-d1, y, 60*a, 30*a)
    b=a
    a=abs(a)
    ellipse(scr, color, (x-d1, y, int(60*abs(a)), int(30*abs(a))))
    ellipse(scr, BLACK, (x-d1, y, int(60*abs(a)), int(30*abs(a))), 1)
    circle(scr, BLUE, (x+c*int(43*a), y+int(12*a)), int(5*a))
    circle(scr, BLACK, (x+int(c*45*a), y+int(15*a)), int(3*a))
    #fins
    polygon(scr, color, [(x, int(y+15*a)), (int(x-c*15*a), int(y+5*a)), (x-int(c*15*a), y+int(25*a))])
    polygon(scr, RED, [(x+c*int(10*a), y+int(20*a)), (x, int(y+30*a)), (int(x+c*15*a), int(y+30*a)), (x+c*int(20*a), y+int(20*a))])
    polygon(scr, RED, [(x+c*int(27*a), y+int(20*a)), (x+c*int(32*a), y+int(30*a)), (x+c*int(37*a), y+int(30*a)), (x+c*int(30*a), y+int(20*a))])
    polygon(scr, RED, [(x+c*int(15*a), y+int(7*a)), (x+c*int(35*a), y+int(7*a)), (x+c*int(30*a), y), (x+c*int(10*a), y)])


#fon
polygon(screen, LIGHT_BLUE, [(0,0), (0, 500),
                              (600, 500), (600, 0)])
polygon(screen, GRAY, [(0,500), (0, 800),
                       (600, 800), (600, 500)])
#halo
circle(screen, YELLOW, (400, 100), 150, 15)
line(screen, (255, 255, 150), (250, 100), (550, 100), 20)
line(screen, (255, 255, 150), (400, 0), (400, 250), 20)
circle(screen, (255, 255, 0), (400, 100), 15, 15)
circle(screen, (255, 255, 0), (265, 100), 15, 15)
circle(screen, (255, 255, 0), (535, 100), 15, 15)
circle(screen, (255, 255, 0), (400, 235), 15, 15)
#prorub`
ellipse(screen, DARK_GRAY, (300, 600, 200, 100))
ellipse(screen, GREEN, (320, 620, 160, 80))

#medved`
#golova
ellipse(screen, GRAY, (50, 400, 100, 50))
ellipse(screen, BLACK, (50, 400,100, 50),1)
circle(screen, BLACK, (100, 415), 3)
line(screen, BLACK, (100, 435), (150, 430), 1)
circle(screen, BLACK, (150, 425), 3)
polygon(screen, GRAY, [(65, 395), (70, 420), (75, 415)])
polygon(screen, BLACK, [(65, 395), (70, 420), (75, 415)], 1)
#body
ellipse(screen, GRAY, (5, 440, 125, 250))
ellipse(screen, BLACK, (5, 440, 125, 250),1)
#leg
ellipse(screen, GRAY, (75, 640, 125, 80))
ellipse(screen, BLACK, (75, 640, 125, 80),1)
#foot
ellipse(screen, GRAY, (150, 690, 80, 50))
ellipse(screen, BLACK, (150, 690, 80, 50),1)
#udochka
line(screen, BLACK, (135, 570), (400,300),3)
line(screen, BLACK, (400,300), (400, 660))
#arm
ellipse(screen, GRAY, (115, 530, 70, 30))
ellipse(screen, BLACK, (115, 530, 70, 30),1)
#fish
ellipse(screen, GREEN, (290, 710, 60, 30))
ellipse(screen, BLACK, (290, 710, 60, 30), 1)
circle(screen, BLUE, (333, 722), 5)
circle(screen, BLACK, (335, 725), 3)
#fins
polygon(screen, GREEN, [(290,725), (275, 715), (275, 735)])
polygon(screen, RED, [(300, 730), (290, 740), (305, 740), (310, 730)])
polygon(screen, RED, [(317, 730), (322, 740), (327, 740), (320, 730)])
polygon(screen, RED, [(305, 717), (325, 717), (320, 710), (300, 710)])

fish(0.5, 340, 725, BLUE, screen)
fish(-0.5, 50, 50, DARK_GRAY, screen)
s=opacity(0.5, 600, 900)
fish(2, 50, 0, RED, s)
screen.blit(s,(50,250))

pygame.display.update()
clock = pygame.time.Clock()
finished = False
while not finished:
    clock.tick(FPS)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            finished = True

pygame.quit()
```
